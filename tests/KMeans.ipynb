```python
"""
Cluster the activation vectors extracted from a trained Convolutional Neural Network.
AUTHOR: ASHWINKUMAR GANESAN.
"""
%matplotlib inline
%load_ext autoreload
%autoreload 2
import pickle
from lib.cluster import NeuralKMeans

# Use Bokeh to plot
from bokeh.plotting import show
from bokeh.io import output_notebook

n_clusters = 10
n_dim = 2
```

```python
# Load the pickled file with CNN layer output values.
cnn_layers = pickle.load(open("data/pickle/cnn_layer/mnist.pickle", 'rb'))
```

```python
# The CNN layers consist of a list of the following items.
# 1. The layer outputs and cost function output.
# 2. The weight and bias terms for each layer.

# Layer 1
layer_0_op = cnn_layers[0][1]
```

```python
# Layer 2
layer_1_op = cnn_layers[0][2]
```

```python
# Layer 3
layer_2_op = cnn_layers[0][3]
```

```python
print(cnn_layers[0][4])
```

```python
cluster_0 = NeuralKMeans(layer_0_op)
cluster_1 = NeuralKMeans(layer_1_op)
cluster_2 = NeuralKMeans(layer_2_op)
```

```python
cluster_0.flatten()
cluster_1.flatten()
cluster_2.flatten()
```

```python
cluster_0.cluster(n_clusters)
cluster_1.cluster(n_clusters)
cluster_2.cluster(n_clusters)
```

```python
cluster_0.tsne(n_dim,0)
cluster_1.tsne(n_dim,0)
cluster_2.tsne(n_dim,0)
```

```python
fig_0 = cluster_0.figure("Activation Matrix Clusters - Layer 0", cnn_layers[0][4])
fig_1 = cluster_1.figure("Activation Matrix Clusters - Layer 1", cnn_layers[0][4])
fig_2 = cluster_2.figure("Activation Matrix Clusters - Layer 2", cnn_layers[0][4])
```

```python
output_notebook()
```
