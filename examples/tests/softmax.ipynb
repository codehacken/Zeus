```python
# Simple implementations using tensorflow.
# MNIST Example with softmax regression.
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
import tensorflow as tf
```

```python
# MNIST dataset has a set of images of size 28x28.
# For a the purpose of a simple softmax implementation, the image is a considered
# to be of vector of size 784.
x = tf.placeholder(tf.float32, [None, 784])
```

```python
# Define the weight vector and the bias term.
W = tf.Variable(tf.zeros([784, 10]))
b = tf.Variable(tf.zeros([10]))
```

```python
# Calculate softmax.
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

```python
# Define a placeholder for output labels.
y_ = tf.placeholder(tf.float32, [None, 10])
```

```python
# Define the cross-entropy.
cross_entropy = -tf.reduce_sum(y_*tf.log(y))
```

```python
# Gradient Descent.
train_step = tf.train.GradientDescentOptimizer(0.0001).minimize(cross_entropy)
```

```python
# Create a session where the computation can be run.
init = tf.initialize_all_variables()
```

```python
# Execute for a 1000 iterations.
with tf.Session() as sess:
    sess.run(init)
    for i in range(1000):
        batch_xs, batch_ys = mnist.train.next_batch(100)
        print(sess.run([cross_entropy, train_step], feed_dict={x: batch_xs, y_: batch_ys}))
        print(sess.run([W, b]))
    
    # Test Predictions.
    correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print(sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
```
